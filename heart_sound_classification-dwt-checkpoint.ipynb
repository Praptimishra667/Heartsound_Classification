```python
import os
import librosa
import numpy as np
import tqdm
import matplotlib.pyplot as plt
```

```python
# Define paths
base_dir = "Data"
class_labels = ['AS', 'MR', 'MS', 'MVP', 'N']

# Function to process all files and return feature vectors and labels
def process_dataset(base_dir, class_labels, sampling_rate=1000):
    feature_list = []
    label_list = []
    
    for label in class_labels:
        folder_path = os.path.join(base_dir, label)
        
        # Iterate over all wav files in the folder
        for file_name in tqdm.tqdm(os.listdir(folder_path)):
            if file_name.endswith('.wav'):
                file_path = os.path.join(folder_path, file_name)
                
                # Load the audio file
                signal, sr = librosa.load(file_path, sr=None)  # Load with original sampling rate
                
                # Resample to the target sampling rate (2 kHz)
                signal = librosa.resample(signal, orig_sr=sr, target_sr=sampling_rate)
                
                # Normalize the signal
                signal = signal / np.max(np.abs(signal))
                
                # Append features and corresponding label
                feature_list.append(signal)
                label_list.append(label)
    
    # Determine the maximum signal length
    max_length = max([len(signal) for signal in feature_list])
    
    # Pad or truncate all signals to the same length
    processed_features = []
    for signal in feature_list:
        if len(signal) < max_length:
            # Pad with zeros
            signal = np.pad(signal, (0, max_length - len(signal)), 'constant')
        elif len(signal) > max_length:
            # Truncate
            signal = signal[:max_length]
        processed_features.append(signal)
    
    # Convert to numpy arrays
    X = np.array(processed_features, dtype=np.float32)
    y = np.array(label_list, dtype=object)
    
    return X, y

# Process dataset
X, y = process_dataset(base_dir, class_labels)
```

```python
import pywt

# Function to perform DWT with Daubechies 14 wavelet at level 4
def apply_dwt(signal, wavelet='coif5', level=5):
    """
    Apply Discrete Wavelet Transform to a single signal using the specified wavelet and level.
    Args:
        signal: Input signal array.
        wavelet: Wavelet type to use for DWT.
        level: Decomposition level.
    Returns:
        coeffs: A list containing wavelet coefficients.
    """
    # Perform multi-level Discrete Wavelet Transform
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    # Concatenate all coefficients into a single feature vector
    dwt_features = np.hstack(coeffs)
    return dwt_features


# Apply DWT to all signals in X
X_dwt = np.array([apply_dwt(signal) for signal in X])
```

```python
X_dwt.shape
```

```python
def standardize_data(X):
    """
    Apply standardization (z-score normalization) to all features across the entire dataset.
    
    Args:
        X: Input feature tensor (n_samples, n_features, n_channels).
        
    Returns:
        X_standardized: Standardized feature tensor with zero mean and unit variance.
    """
    # Calculate the mean and standard deviation across the entire dataset
    X_mean = np.mean(X, axis=(0, 1), keepdims=True)  # Mean across all samples and features for each channel
    X_std = np.std(X, axis=(0, 1), keepdims=True)  # Standard deviation across all samples and features for each channel
    
    # Avoid division by zero
    X_std = np.where(X_std == 0, 1e-10, X_std)

    # Apply z-score normalization
    X_standardized = (X - X_mean) / X_std
    return X_standardized


# Standardize the DWT features
X_dwt_standardized = standardize_data(X_dwt)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming X_stft is a 3D array (1000, frequency_bins, time_frames)
# and y is a 1D array (1000,) containing the labels

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X_dwt_standardized, y_enc, test_size=0.2, random_state=42)

# Check the shapes of the resulting datasets
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout

# Reshape input for CNNs
# Adding an additional dimension for CNN (timesteps, features)
X_train_reshape = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_reshape = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

## Simple CNN model

```python
# Define the CNN model
model = Sequential([
    Conv1D(32, kernel_size=2, activation='relu', input_shape=(4135, 1)),  # Corrected input shape
    MaxPooling1D(pool_size=1),
    
    Conv1D(64, kernel_size=2, activation='relu'),
    MaxPooling1D(pool_size=1),
    
    Flatten(),
    
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # multiclass classification with softmax activation
])
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the CNN model
history = model.fit(X_train_reshape, y_train, epochs=10, batch_size=32, validation_data=(X_test_reshape, y_test))

# # Evaluate CNN
cnn_accuracy = model.evaluate(X_test_reshape, y_test, verbose=0)
print(f"CNN Accuracy: {cnn_accuracy[1] * 100:.2f}%")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Reshape, GRU, Dense, Dropout

# Define the CNN-GRU model
model = Sequential([
    # CNN Layers
    Conv1D(32, kernel_size=2, activation='relu', input_shape=(4135, 1)),  # Input shape (4135, 1)
    MaxPooling1D(pool_size=2),
    
    Conv1D(64, kernel_size=2, activation='relu'),
    MaxPooling1D(pool_size=2),
    
    Flatten(),
    
    # Reshape for GRU layer
    Reshape((64, -1)),  # Reshape to (timesteps, features). Ensure the feature dimension matches.
    
    # GRU Layer
    GRU(64, return_sequences=False),
    Dropout(0.5),
    
    # Fully Connected Layers
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # Output for 5 classes
])
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the CNN model
history = model.fit(X_train_reshape, y_train, epochs=50, batch_size=32, validation_data=(X_test_reshape, y_test))

# # Evaluate CNN
cnn_accuracy = model.evaluate(X_test_reshape, y_test, verbose=0)
print(f"CNN Accuracy: {cnn_accuracy[1] * 100:.2f}%")
```
