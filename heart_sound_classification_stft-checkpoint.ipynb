```python
import os
import librosa
import numpy as np
import tqdm
import matplotlib.pyplot as plt
```

```python
# Define paths
base_dir = "Data"
class_labels = ['AS', 'MR', 'MS', 'MVP', 'N']

# Function to process all files and return feature vectors and labels
def process_dataset(base_dir, class_labels, sampling_rate=1000):
    feature_list = []
    label_list = []
    
    for label in class_labels:
        folder_path = os.path.join(base_dir, label)
        
        # Iterate over all wav files in the folder
        for file_name in tqdm.tqdm(os.listdir(folder_path)):
            if file_name.endswith('.wav'):
                file_path = os.path.join(folder_path, file_name)
                
                # Load the audio file
                signal, sr = librosa.load(file_path, sr=None)  # Load with original sampling rate
                
                # Resample to the target sampling rate (2 kHz)
                signal = librosa.resample(signal, orig_sr = sr, target_sr = sampling_rate)
                
                # Normalize the signal
                signal = signal / np.max(np.abs(signal))
                
                # Append features and corresponding label
                feature_list.append(signal)
                label_list.append(label)
    
    # Convert to numpy arrays
    X = np.array(feature_list,dtype=object)
    y = np.array(label_list,dtype=object)
    
    return X, y

# Process dataset
X, y = process_dataset(base_dir, class_labels)
```

```python
# Get unique elements and create a mapping to integers
unique_elements = np.unique(y)
mapping = {elem: idx for idx, elem in enumerate(unique_elements)}

# Map the original array to integers
y_enc = np.vectorize(mapping.get)(y)
```

```python
import pandas as pd

# Assuming y_train is a NumPy array or a pandas Series
y_onehot = pd.get_dummies(y_enc).values
y_onehot.shape
```

```python
plt.plot(X[0])
```

```python
n_fft = 512
hop_length = 256
max_time_frames = 0  # To track the maximum number of time frames

# First, calculate the STFTs and track the maximum time frames length
X_stft = []

for signal in tqdm.tqdm(X):
    stft_result = librosa.stft(signal, n_fft=n_fft, hop_length=hop_length)
    X_stft.append(np.abs(stft_result))  # Store magnitude of STFT
    
    # Track maximum number of time frames
    max_time_frames = max(max_time_frames, stft_result.shape[1])

# Now, pad or truncate each STFT to have consistent time frames length
padded_X_stft = []

for stft_result in X_stft:
    # Pad the time dimension if necessary (or truncate)
    padded_result = np.pad(stft_result, ((0, 0), (0, max_time_frames - stft_result.shape[1])), mode='constant')
    padded_X_stft.append(padded_result)

# Convert list of STFT results to a numpy array
X_stft = np.array(padded_X_stft)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming X_stft is a 3D array (1000, frequency_bins, time_frames)
# and y is a 1D array (1000,) containing the labels

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X_stft, y_onehot, test_size=0.2, random_state=42)

# Check the shapes of the resulting datasets
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# For CNN: Keep the 3D shape (samples, height, width, channels)
X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2], 1)  # (800, 257, 16, 1)
X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2], 1)      # (200, 257, 16, 1)
```

```python
X_train_cnn.shape
```

```python
# --- CNN Model ---
cnn_model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(X_train_cnn.shape[1], X_train_cnn.shape[2], 1)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(5, activation='softmax')  # 5 classes for classification
])
```

```python
cnn_model.summary()
```

```python
# Compile the model
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the CNN model
history = cnn_model.fit(X_train_cnn, y_train, epochs=50, batch_size=32, validation_data=(X_test_cnn, y_test))

# # Evaluate CNN
cnn_accuracy = cnn_model.evaluate(X_test_cnn, y_test, verbose=0)
print(f"CNN Accuracy: {cnn_accuracy[1] * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Assuming `history` contains the training history from model.fit()

# Extract loss and accuracy
epochs = range(1, len(history.history['loss']) + 1)
loss = history.history['loss']
val_loss = history.history['val_loss']
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Plot Loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, label='Training Loss', marker='o')
plt.plot(epochs, val_loss, label='Validation Loss', marker='x')
plt.title('Epoch vs Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
# plt.grid()

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, accuracy, label='Training Accuracy', marker='o')
plt.plot(epochs, val_accuracy, label='Validation Accuracy', marker='x')
plt.title('Epoch vs Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
# plt.grid()

# Show plots
plt.tight_layout()
plt.show()
```

## CNN with AttentionLayer

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Define the Attention Layer
class AttentionLayer(layers.Layer):
    def __init__(self):
        super(AttentionLayer, self).__init__()

    def build(self, input_shape):
        # Create the attention weights (query, key, value matrices)
        self.query = self.add_weight(name="query", shape=(input_shape[-1], input_shape[-1]), initializer="random_normal")
        self.key = self.add_weight(name="key", shape=(input_shape[-1], input_shape[-1]), initializer="random_normal")
        self.value = self.add_weight(name="value", shape=(input_shape[-1], input_shape[-1]), initializer="random_normal")

    def call(self, inputs):
        # Calculate attention scores (query * key)
        query_out = tf.matmul(inputs, self.query)
        key_out = tf.matmul(inputs, self.key)
        value_out = tf.matmul(inputs, self.value)
        
        # Calculate attention weights using softmax
        attention_weights = tf.nn.softmax(tf.matmul(query_out, key_out, transpose_b=True), axis=-1)
        
        # Apply attention weights to values
        attention_output = tf.matmul(attention_weights, value_out)
        
        return attention_output

# --- CNN with Attention Model ---
cnn_attention_model = models.Sequential([
    # First convolutional layer
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(X_train_cnn.shape[1], X_train_cnn.shape[2], 1)),
    layers.MaxPooling2D((2, 2)),

    # Second convolutional layer
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Attention mechanism
    AttentionLayer(),

    # Flatten the output to feed into dense layers
    layers.Flatten(),
    
    # Fully connected layer
    layers.Dense(64, activation='relu'),
    
    # Output layer (5 classes)
    layers.Dense(5, activation='softmax')  # 5 classes for classification
])
```

```python
cnn_attention_model.summary()
```

```python
# Compile the model
cnn_attention_model.compile(optimizer='adam',
                            loss='categorical_crossentropy',
                            metrics=['accuracy'])

# Train the CNN with Attention model
history = cnn_attention_model.fit(X_train_cnn, y_train, epochs=50, batch_size=32, validation_data=(X_test_cnn, y_test))

# Evaluate the CNN with Attention model
cnn_attention_accuracy = cnn_attention_model.evaluate(X_test_cnn, y_test, verbose=0)
print(f"CNN with Attention Accuracy: {cnn_attention_accuracy[1] * 100:.2f}%")
```
